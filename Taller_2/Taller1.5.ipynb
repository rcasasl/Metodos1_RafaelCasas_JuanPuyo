```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import sympy as sym
```

```python
#Show that the D4f operator is given by:
#Esto se hara usando la funcion seno y demostrando que la ecuacion dada equivale a la ecuacion real de la cuarta derivada, comparando sus graficas
x = np.linspace(0,2*np.pi,50)
h = x[1]-x[0]
```

```python
sin = lambda x: np.sin(x)
```

```python
f = sin(x)
```

```python
f
```

```python
def CuartaDerivative(x,f,h):
    return ((f(x+2)-4*f(x+1)+6*f(x)-4*f(x-1)+f(x-2))/(h**4))
```

```python
CuartaDerivada = CuartaDerivative(x,sin,h)
```

```python
fig =plt.figure(figsize=(7,3))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)


ax.scatter(x,np.sin(x))
ax1.scatter(x,CuartaDerivada,Color='red')
```
