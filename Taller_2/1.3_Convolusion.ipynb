```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import sympy as sym
```

```python
x = np.linspace(-10,10,400)
h = 0.05
funcion = lambda x: (1)/((np.exp((-x)**2))**(1/2))
val_der = np.zeros(400)
```

```python
M = [1,0,-1]
def Operador(x,h):
    derivada = (1/(2*h))*(M[0]*funcion(x+h)+ M[1]*funcion(x)+M[2]*funcion(x-h))
    return derivada
```

```python
val_der[:]=Operador(x,h)
```

```python
val_der
```

```python
fig =plt.figure(figsize=(7,5))
ax = fig.add_subplot(2,2,1)

ax.scatter(x,val_der)
```
