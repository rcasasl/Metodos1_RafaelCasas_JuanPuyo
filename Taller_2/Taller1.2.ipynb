```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import sympy as sym
```

```python
#Usando la defnicion de derivada central (con h = 0:05) estime la derivada de la funcion:
#def Function(x):
 #   return (1)/((np.exp((-x)**2))**(1/2))
    
x = sym.Symbol('x',Real=True)
func = 1/((1+sym.exp(-(x)**2))**(1/2))
```

```python
func
```

```python
f = sym.lambdify([x],func,'numpy')
```

```python
f(0.3)
```

```python
#a) En el intervalo -10<x<10.
x = np.linspace(-10,10,400)
y = f(x)
h = 0.05
```

```python
#Function(x)
plt.scatter(x,y)
```

```python
f
```

```python
def DerivativeC(x,f,h):
    return ((f(x+h)-f(x-h))/(2*h))
```

```python
def Derivative(x,f,h):
    return ((f(x+h)-f(x))/h)
```

```python
DerivadaC = DerivativeC(x,function,h)
Derivada = Derivative(x,function,h)
```

```python
DerivadaC
```

```python
Derivada
```

```python
error = np.abs(DerivadaC-Derivada)
```

```python
error
```

```python
fig =plt.figure(figsize=(7,5))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)

ax.plot(x,Derivada)
ax1.plot(x,DerivadaC,color='red')
ax2.plot(x,error)
```
