```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import sympy as sym
```

```python
#Demuestre la formula alternativa para la estimacion de la segunda derivada discreta:
#Para demostrarla voy a usar una funcion seno la cual ya conozco su segunda derivada, y comparar su grafica con la grafica generada por la ecuacion dada

x = np.linspace(0,2*np.pi,50)
h = x[1]-x[0]
```

```python
sin = lambda x: np.sin(x)
```

```python
f = sin(x)
```

```python
f
```

```python
def SegundaDerivadaDisc(x,f,h):
    return ((f(x+2)-2*f(x)+f(x-2))/(4*(h**2)))
```

```python
def SegundaDerivada(x,f,h):
    return ((f(x+1)-2*f(x)+f(x-1))/(h**2))
```

```python
SegundaDerivadaF = SegundaDerivadaDisc(x,sin,h)
SegundaDerivadaR = SegundaDerivada(x,sin,h)
```

```python
SegundaDerivadaF
```

```python
fig =plt.figure(figsize=(7,3))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)


ax.scatter(x,SegundaDerivadaF)
ax1.scatter(x,SegundaDerivadaR,Color='red')
```
