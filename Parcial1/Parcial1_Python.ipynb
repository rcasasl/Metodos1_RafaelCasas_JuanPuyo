```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = np.linspace(0.1,1.1,100)
```

```python
tan = lambda x: (np.tan(x))**0.5
```

```python
#Punto c
def DerivativeP(f,x,h=0.01):
    return ((-3)*f(x)+4*f(x+h)-f(x+2*h))/(2*h)
```

```python
#Punto d
def DerivativeC(f,x,h=0.01):
    return ((f(x+h)-f(x-h))/(2*h))
```

```python
#Punto e
#La derivada analítica de la función f(x)=sqrt(tanx) es:
    
    #f'(x)= sec^2(x)/(2*sqrt(tan(x)))

df= lambda x: (np.cos(x))**-2/(2*np.sqrt(np.tan(x)))

DerivativeR=df(x)
```

```python
DerivadaP = DerivativeP(tan,x,h) #Roja
DerivadaC = DerivativeC(tan,x,h) #Azul
DerivadaR = DerivativeR #Negra
```

```python
#Punto f
error1 = np.abs(DerivadaP-DerivadaR)
error2 = np.abs(DerivadaC - DerivadaR)
```

```python
#Punto f
fig =plt.figure(figsize=(30,30))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)

ax.scatter(x,DerivadaP,color='red')
ax1.scatter(x,DerivadaC)
ax2.scatter(x,DerivadaR,color='k')
ax3.scatter(x,error1,color='red')
ax3.scatter(x,error2)

"""
En la grafica del error, la funcion en azul equivale al error nodal de la Derivada Central, mientras que la funcion roja equivale al error nodal de la Derivada Progresiva.

Las graficas no son igual de correctas, puesto que en la derivada progresiva se esta acercando a el punto x porgresivamente, desde adelante. En comparacion con la derivada central donde se hace un promedio \
entre la derivada progresiva y regresiva, por lo que se disminuye el error.
"""
```
